```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encode=LabelEncoder()
```

```python
encoded_state = label_encode.fit_transform(crop['STATE']).astype('int32')
crop.loc[:, 'STATE'] = encoded_state
```

```python
X=crop.drop(['CROP'],axis=1)
y=crop['CROP']
```

```python
crop
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3)
```

```python
X_train
```

```python
#USING DECISION TREE
from sklearn.metrics import accuracy_score
from sklearn import tree
decision_tree=tree.DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
```

```python
y_pred=decision_tree.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
pickle.dump(decision_tree, open('crop_model.pkl', 'wb'))
pickle.dump(label_encode, open('label_encode.pkl', 'wb'))
```
